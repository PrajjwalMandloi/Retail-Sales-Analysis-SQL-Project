```python
from faker import Faker
import pandas as pd
import random

fake = Faker('en_IN')

# CUSTOMERS

customers = []

for customer_id in range(1, 51):
    customers.append([
        customer_id,
        fake.name(),
        random.choice(['Male', 'Female']),
        random.randint(18, 60),
        fake.city()
    ])

customers_df = pd.DataFrame(
    customers,
    columns=[
        'customer_id',
        'customer_name',
        'gender',
        'age',
        'city'
    ]
)


# PRODUCTS

products_list = [
    ('Laptop', 'Electronics'),
    ('Mobile', 'Electronics'),
    ('Tablet', 'Electronics'),
    ('Smart Watch', 'Electronics'),
    ('Earbuds', 'Electronics'),
    ('Keyboard', 'Accessories'),
    ('Mouse', 'Accessories'),
    ('Monitor', 'Accessories'),
    ('Webcam', 'Accessories'),
    ('Speaker', 'Accessories'),
    ('Printer', 'Accessories'),
    ('Power Bank', 'Accessories'),
    ('Mixer', 'Home Appliances'),
    ('Iron', 'Home Appliances'),
    ('Vacuum Cleaner', 'Home Appliances'),
    ('Air Purifier', 'Home Appliances'),
    ('Water Purifier', 'Home Appliances'),
    ('Microwave', 'Home Appliances'),
    ('Refrigerator', 'Home Appliances'),
    ('Washing Machine', 'Home Appliances'),
    ('Office Chair', 'Furniture'),
    ('Study Table', 'Furniture'),
    ('Bookshelf', 'Furniture'),
    ('Sofa', 'Furniture'),
    ('Dining Table', 'Furniture'),
    ('LED TV', 'Electronics'),
    ('Camera', 'Electronics'),
    ('Gaming Console', 'Electronics'),
    ('Router', 'Electronics'),
    ('Projector', 'Electronics')
]

products = []

for product_id, (name, category) in enumerate(products_list, start=1):
    products.append([
        product_id,
        name,
        category,
        random.randint(500, 80000)
    ])

products_df = pd.DataFrame(
    products,
    columns=[
        'product_id',
        'product_name',
        'category',
        'price'
    ]
)


# ORDERS

orders = []

for order_id in range(1, 201):
    orders.append([
        order_id,
        random.randint(1, 50),
        fake.date_between(
            start_date='-1y',
            end_date='today'
        )
    ])

orders_df = pd.DataFrame(
    orders,
    columns=[
        'order_id',
        'customer_id',
        'order_date'
    ]
)


# ORDER ITEMS

order_items = []

order_item_id = 1

for order_id in range(1, 201):

    num_products = random.randint(1, 4)

    selected_products = random.sample(
        range(1, 31),
        num_products
    )

    for product_id in selected_products:

        order_items.append([
            order_item_id,
            order_id,
            product_id,
            random.randint(1, 5)
        ])

        order_item_id += 1

order_items_df = pd.DataFrame(
    order_items,
    columns=[
        'order_item_id',
        'order_id',
        'product_id',
        'quantity'
    ]
)


# SAVE CSV FILES

customers_df.to_csv('customers.csv', index=False)
products_df.to_csv('products.csv', index=False)
orders_df.to_csv('orders.csv', index=False)
order_items_df.to_csv('order_items.csv', index=False)

print("Files Generated Successfully!")
```
